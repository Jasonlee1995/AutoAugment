```python
import augmentation, torch, torchvision
import matplotlib.pyplot as plt

# Ignore Warning
import warnings
warnings.filterwarnings(action='ignore')
```

```python
dataset = torchvision.datasets.CIFAR10('../../../../data/OCR/Jason/cifar10/', train=False, download=False)
```

```python
ex1, ex2, ex3, ex4 = dataset[2], dataset[25], dataset[28], dataset[42]
```

```python
aug_list = [[augmentation.Rotate(1, -20, minus=False), augmentation.Rotate(1, 20, minus=False)], 
            [augmentation.TranslateX(1, -10, minus=False), augmentation.TranslateX(1, 10, minus=False)],
            [augmentation.TranslateY(1, -10, minus=False), augmentation.TranslateY(1, 10, minus=False)],
            [augmentation.ShearX(1, -0.2, minus=False), augmentation.ShearX(1, 0.2, minus=False)],
            [augmentation.ShearY(1, -0.2, minus=False), augmentation.ShearY(1, 0.2, minus=False)]]

title_list = [['Rotate -20', 'Rotate 20'],
              ['TranslateX -10', 'TranslateX 10'],
              ['TranslateY -10', 'TranslateY 10'],
              ['ShearX -0.2', 'ShearX 0.2'],
              ['ShearY -0.2', 'ShearY 0.2']]
```

```python
plt.figure(figsize=(15, 45))
plt.rcParams['axes.titlesize'] = 20
plt.axis('off')

for i in range(len(aug_list)):
    plt.subplot(8, 3, 3*i+1)
    image, label = ex1
    img = plt.imshow(image); img.axes.get_xaxis().set_visible(False); img.axes.get_yaxis().set_visible(False)
    plt.gca().set_title('GT')
    
    for j in range(len(aug_list[i])):
        plt.subplot(8, 3, 3*i + j + 2)
        aug_image = aug_list[i][j](image)
        aug_image = plt.imshow(aug_image)
        aug_image.axes.get_xaxis().set_visible(False); aug_image.axes.get_yaxis().set_visible(False)
        plt.gca().set_title(title_list[i][j])
```
